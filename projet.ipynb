```python
import requests

API_KEY = "258275c0180041fb82b5f9d574799473" # CREATE YOUR OWN API KEY

url = f" https://newsapi.org/v2/everything?q=karate&from=2023-03-20&sortBy=publishedAt&apiKey={API_KEY} " # SEARCH THE PROPER URL INSIDE THE DOCUMENTATION

response = requests.get(url)

if response.status_code == 200:
    data = response.json()
    print(data)
else:
    print('Error:', response.status_code)
```
